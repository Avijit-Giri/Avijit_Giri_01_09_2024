```python
from sklearn.impute import SimpleImputer

# Impute missing values with the mean of the respective columns
imputer = SimpleImputer(strategy='mean')

# Apply the imputer to the relevant columns
df[['Targeted Productivity', 'Overtime', 'No. of Workers', 'Actual Productivity']] = imputer.fit_transform(
    df[['Targeted Productivity', 'Overtime', 'No. of Workers', 'Actual Productivity']]
)

# Verify that there are no missing values left
df[['Targeted Productivity', 'Overtime', 'No. of Workers', 'Actual Productivity']].isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

#  Standardize the features
features = ['Targeted Productivity', 'Overtime', 'No. of Workers', 'Actual Productivity']
x = df[features].values
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

#  Perform PCA
pca = PCA()
principal_components = pca.fit_transform(x_scaled)

# Explained Variance Ratio
explained_variance = pca.explained_variance_ratio_

#  Calculate cumulative explained variance
cumulative_variance = np.cumsum(explained_variance)

# Find the number of components that explain at least 90% of the variance
num_components_90_var = np.argmax(cumulative_variance >= 0.90) + 1

# Display results
num_components_90_var, explained_variance, cumulative_variance
```

```python
#  Plot the explained variance by each principal component
plt.figure(figsize=(8, 6))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.5, align='center', label='Individual Explained Variance')
plt.step(range(1, len(cumulative_variance) + 1), cumulative_variance, where='mid', label='Cumulative Explained Variance')
plt.ylabel('Explained Variance Ratio')
plt.xlabel('Principal Components')
plt.title('Explained Variance by Each Principal Component')
plt.legend(loc='best')
plt.show()
```

Principal Components and Explained Variance:

The first principal component (PC1) might explain the most variance in the dataset, followed by the second, third, etc.
If PC1 explains a significant amount of variance (e.g., 70%), it suggests that the most variability in productivity data is along this component's direction.
Number of Components Needed:

If 2 components explain over 90% of the variance, we could effectively reduce our analysis to these 2 dimensions, simplifying further analysis and visualization.
Understanding Productivity Dynamics:

Each principal component is a linear combination of the original features:
PC1 could represent a combination of "Targeted Productivity" and "Actual Productivity" if these have the highest weights (loadings).
PC2 could be influenced more by "Overtime" and "No. of Workers."
By examining the loadings of each principal component, we can determine which factors most significantly impact productivity and how they interact.
Applications:

Dimensionality Reduction: Reducing the number of variables while retaining most of the variance.
Visualization: Use the first two or three components for 2D or 3D visualization.
Feature Engineering: New principal components can be used as input features for predictive modeling.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error



# Step 1: Data Preparation
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Ensure 'Actual Productivity' is numeric
df['Actual Productivity'] = pd.to_numeric(df['Actual Productivity'], errors='coerce')

# Drop any rows with missing values in 'Actual Productivity'
df.dropna(subset=['Actual Productivity'], inplace=True)

# Step 2: ARIMA Model Building
# Fit ARIMA model to the Actual Productivity data
# Here, we'll start with an (p=1, d=1, q=1) model, you can fine-tune based on AIC/BIC values
model = ARIMA(df['Actual Productivity'], order=(1, 1, 1))
model_fit = model.fit()

# Forecast for the next 4 periods (assuming these are quarters or weeks based on your dataset)
forecast = model_fit.forecast(steps=4)

# Step 3: Model Evaluation
# Assume we have some actual data to compare with for evaluation (let's create a mock actual data for illustration)
actual = df['Actual Productivity'][-4:]  # Last 4 periods as "actual" data for simplicity

# Compute MAPE and MSE
mape = mean_absolute_percentage_error(actual, forecast)
mse = mean_squared_error(actual, forecast)

print(f"MAPE: {mape:.2f}")
print(f"MSE: {mse:.2f}")

# Step 4: Visualization
plt.figure(figsize=(10, 6))
plt.plot(df['Actual Productivity'], label='Actual')
plt.plot(pd.date_range(start=df.index[-1], periods=5, freq='Q')[1:], forecast, label='Forecast', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Actual Productivity')
plt.title('Actual vs Forecasted Productivity')
plt.legend()
plt.show()

# Step 5: Interpretation
# Provide interpretation of the results
```

Interpretation of the Results
MAPE and MSE: These metrics provide insights into the model's accuracy:

MAPE: A lower value indicates better predictive accuracy.
MSE: A lower value indicates less deviation of the predictions from the actual values.
Forecast Plot: The plot shows how well the ARIMA model's forecast aligns with actual productivity values. The closer the forecast line to the actual data, the more accurate the model.

Model Fine-Tuning: If the model's performance is not satisfactory, consider adjusting the p, d, q parameters or transforming the data further (e.g., seasonal decomposition, differencing).

```python

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler



# Step 1: Data Preparation
# Select the relevant columns
features = ['Actual Productivity', 'Overtime', 'No. of Workers']
df_selected = df[features].copy()

# Standardize the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_selected)

# Step 2: K-Means Clustering
# Initialize variables for the Elbow method
sse = []  # Sum of squared distances for each k

# Testing for different numbers of clusters
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df_scaled)
    sse.append(kmeans.inertia_)

# Step 3: Determine Optimal Number of Clusters using Elbow Method
plt.figure(figsize=(8, 5))
plt.plot(k_range, sse, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Distances')
plt.show()

# Let's assume the optimal number of clusters is chosen (e.g., 3) based on the Elbow plot
optimal_k = 3  # Set this to the optimal k based on the Elbow plot

# Apply KMeans with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
df['Cluster'] = kmeans.fit_predict(df_scaled)

# Step 4: Visualization
# Plot the clusters
plt.figure(figsize=(10, 6))
plt.scatter(df['Overtime'], df['Actual Productivity'], c=df['Cluster'], cmap='viridis', marker='o', edgecolor='k')
plt.title('K-Means Clustering of Productivity Data')
plt.xlabel('Overtime')
plt.ylabel('Actual Productivity')
plt.colorbar(label='Cluster')
plt.show()

# Step 5: Interpretation
# Print cluster centroids and interpret the clusters
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
print(f"Cluster Centers:\n{centroids}")

# Interpretation: Analyze how different clusters differ in terms of 'Overtime', 'Actual Productivity', and 'No. of Workers'
```

Interpretation of the Results
Elbow Method: The plot shows the sum of squared distances for different values of k. The optimal k is at the "elbow" point, which indicates the best trade-off between complexity and model performance.

Cluster Visualization: The scatter plot visualizes the clusters. Each color represents a different cluster. By examining the plot, we can identify how different groups of workers contribute to productivity levels.

Cluster Centers: The cluster centers provide information on the average Overtime, Actual Productivity, and No. of Workers for each cluster. These values help us understand the characteristics of each cluster.

Interpretation of Clusters:

Clusters can reveal different segments of worker behavior. For example, one cluster might represent workers with high overtime but lower productivity, while another represents workers with balanced overtime and high productivity.
Understanding these segments can help in planning resource allocation, optimizing worker schedules, and enhancing overall productivity.

```python


# Step 1: Data Preparation
# Assuming 'Department Quarterly Spend' is a column in the dataset or needs to be calculated

# If 'Department Quarterly Spend' is not in the dataset, we assume each department spends equally to sum up to Rs. 8.4 lakh per quarter
total_quarterly_budget = 8.4 * 1e5  # Rs. 8.4 lakh
num_departments = df['Department'].nunique()
equal_quarterly_spend = total_quarterly_budget / num_departments

# Assign equal spend to each department
df['Department Quarterly Spend'] = df['Department'].map(df['Department'].value_counts()) * equal_quarterly_spend / df['Department'].map(df['Department'].value_counts()).sum()

# Step 2: Calculate Department Value
df['Department Value'] = df['Actual Productivity'] / df['Department Quarterly Spend']

# Step 3: Analyze the Department Values
department_values = df.groupby('Department')['Department Value'].mean()
print(department_values)

# Step 4: Suggest a Revised Budget Allocation Strategy
# Increase allocation to departments with higher values and decrease for lower values
total_budget = 8.4 * 1e5
budget_allocation = department_values / department_values.sum() * total_budget

print("\nRevised Budget Allocation Strategy (in Rs):")
print(budget_allocation)

# Step 5: Prepare Data for PowerPoint Presentation
# Save the results to a CSV for easier visualization in PowerPoint
budget_allocation.to_csv('AssignmentData.csv', index=True)

# At this point, you would manually create the PowerPoint slides based on these outputs.
```
